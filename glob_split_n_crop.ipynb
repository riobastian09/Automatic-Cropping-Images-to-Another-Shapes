```python
import numpy as np
import cv2, os, glob
import matplotlib.pyplot as plt
from os.path import splitext, basename, exists, join, normpath, dirname
```

```python
def image_resize(image, width = None, height = None, inter = cv2.INTER_AREA):
    dim = None
    (h, w) = image.shape[:2]

    if width is None and height is None:
        return image

    if width is None:
        r = height / float(h)
        dim = (int(w * r), height)

    else:
        r = width / float(w)
        dim = (width, int(h * r))

    resized = cv2.resize(image, dim, interpolation = inter)

    return resized
```

```python
points = np.array([[310, 0], [310, 828], [1450, 828], [1450, 0]]) 
points = points.reshape((len(points),1,2))

for img_path in sorted(glob.glob(r"./*/*.jpg")):
    name = ''
    base_name = ''
    img = 0
    blackout = 0
    final = 0

    img = cv2.imread(img_path)

    img_name  = basename(normpath(img_path))
    base_name = dirname(img_path)
    locate = "Crop "+str(basename(base_name))

    if exists(join(base_name, locate)):
        print(f'{locate} already exists, skipping...')
    else:
        print(f'Creating Folder : {join(base_name, locate)}')
        os.makedirs(join(base_name, locate))
    
    print('Processing',base_name)

    blackout = cv2.fillPoly(np.zeros_like(img, dtype=np.uint8), [points], (255, 255, 255))

    x, y, w, h = cv2.boundingRect(points)
    img  =  img[y:y+h, x:x+w]
    
    blackout  =  blackout[y:y+h, x:x+w]

    blackout = cv2.cvtColor(blackout, cv2.COLOR_BGR2GRAY)
    dataAnd = cv2.bitwise_and(img, img, mask=blackout)
    final = image_resize(dataAnd, width=800, height= 600)

    black_background = np.ones((600,800, 3), dtype = int)
    black_background[10:590, :] = final

    name = join(join(base_name, locate), "Cropped-" + basename(img_name))
    print ('Creating : ' + name, end='\r', flush=True)
    

    print(f'\nProcessing {name} Done!\n' if cv2.imwrite(name, black_background) else "Invalid image!\n")
```
